## Reestruturação inicial dos dados

Neste primeiro momento, o foco inicial do trabalho foi de importar e realizar o primeiro contato com o dataset para compreender as informações de trabalho recebidas.

O primeiro passo que eu realizei, foi o de reestruturar o dataset, pois as informações estavam organizadas em dicionários dentro das colunas do dataset, impossibilitando possíveis tratamentos.

```python
import pandas as pd
import numpy as np
```

```python
# importando os dados do API
dados = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
dados
```

```python
# separando os dados que estão dentro dos dicts das colunas(customer,  phone, internet, account)
dados_customer = pd.json_normalize(data=dados.customer)
dados_customer
```

```python
# separando os dados que estão dentro dos dicts das colunas(customer,  phone, internet, account)
dados_phone = pd.json_normalize(data=dados.phone)
dados_phone
```

```python
# separando os dados que estão dentro dos dicts das colunas(customer,  phone, internet, account)
dados_internet = pd.json_normalize(data=dados.internet)
dados_internet
```

```python
# separando os dados que estão dentro dos dicts das colunas(customer,  phone, internet, account)
dados_account = pd.json_normalize(data=dados.account)
dados_account
```

```python
# usando função concat pra reagrupar as informações e deixar elas em função do customer ID
# pesquisar sobre o merge e aplicar ele
dataset = pd.concat([dados, dados_customer, dados_phone, dados_internet, dados_account], axis=1)
dataset
```

```python
# removendo as colunas que estavam com os dicts
dataset.drop(columns=['customer', 'phone', 'internet', 'account'], inplace=True)
dataset
```

```python
pd.set_option("display.max_columns", 1000)
```

```python
# resultado final após concatenação das colunas novas
dataset
```

## verificação de não conformidades

Nessa etapa eu analisei e verifiquei possíveis inconsistências nos dados(dados faltantes, dados NaN ou invalidos, etc...) que fossem necessários de tratar.

```python
# verificação de algumas informações do dataset
dataset.info()
```

```python
# isolando uma coluna
dataset.iloc[:, [1]]
```

```python
# agrupando pra verificar quantos IDs tem em cada identificação do status do churn
# identificação de inconsistencia: entradas de churn vazia(sem identificação de yes ou no)
dataset.groupby('Churn').nunique()
```

```python
# identificação de inconsistencia: ao inves de 0 e 1, deveria estar escrito yes(1) e no(0). objetivo da mudança seria para padronizar os valores
# de acordo com o resto do dataset para melhor consistencia
dataset.groupby('SeniorCitizen').nunique()
```

```python
# inconsistencia nos dados: o tenure = 0 possui uma quantidade distoante de elementos se comparado com o resto da serie, o que pode comprometer sua
# visualização gráfica e tratamento futuro
dataset.groupby('tenure').nunique()
```

```python
# inconsistencia nos dados: o charges.total vazio pode comprometer sua visualização gráfica e realização de calculos
dataset.groupby('Charges.Total').nunique()
```

```python
dataset_churn = dataset.iloc[:, [1]]
dataset_churn
```

```python
# aplicação do metodo to_numeric pra transformar o charges.total em float
dataset['Charges.Total'] = pd.to_numeric(dataset['Charges.Total'], errors='coerce')
dataset.info()
```

```python
# verificação das entradas com valores NaN
dataset[dataset['Charges.Total'].isnull()]
```

```python
# aplicação do dropna para remover as entradas NaN
dataset.dropna(subset = ['Charges.Total'], inplace = True)
```

```python
# nova verificação da mudança
dataset.info()
```

```python
# inconsistencia do tenure foi resolvida com o tratamento do charges.total, indicando que o problema das duas variáveis tinham a mesma causa
dataset.groupby('tenure').nunique()
```

```python
# tratamento das entradas nulas do churn
dataset = dataset.query('Churn == "Yes" or Churn == "No"')
dataset
```

```python
# reset do index após tratamento final do churn
dataset = dataset.reset_index(drop=True)
dataset
```
